```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
chicken = cv2.imread("./img/chicken.png")
chicken = cv2.cvtColor(chicken, cv2.COLOR_BGR2GRAY)
plt.imshow(chicken, cmap='gray')
```

```python
chicken_hist = cv2.equalizeHist(chicken)
plt.imshow(chicken_hist, cmap='gray')
```
